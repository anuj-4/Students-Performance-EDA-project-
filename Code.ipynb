# **Student** **Performance Project**

```python
import pandas as pd
import numpy as np
```

```python
import plotly.express as px
```

```python
import plotly.graph_objects as go
```

```python
df = pd.read_excel('/content/student_performance_iris_style_300.xlsx')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.set_index('student_id', inplace=True)
display(df.head())
```

```python
df = df.reset_index()
display(df.head())
```

```python
df.info()
```

```python
total_student=df.shape[0]
total_student
```

```python
avg_study_hour=df['study_hours'].mean()
float(avg_study_hour)
```

```python
avg_attendance=df['attendance_pct'].mean()
float(avg_attendance)
```

```python
avg_practice_score = df['practice_score'].mean()
float(avg_practice_score)
```

```python
top_performance=df['performance_level'].value_counts().idxmax()
```

```python
kpi_table = pd.DataFrame({
    'Total Students': [total_student],
    'Average Study Hours': [round(avg_study_hour,2)],
    'Average Attendance (%)': [round(avg_attendance,2)],
    'Average Practice Score': [round(avg_practice_score,2)],
    'Most Frequent Performance Level': [top_performance]
})

kpi_table
```

```python
fig1= px.scatter(df, x='study_hours', y='practice_score',
                 color='performance_level',
                 title='Scatter Plot of Study Hours vs Practice Score')

fig1.show()
```

```python
fig2= px.scatter(df, x='attendance_pct', y='practice_score',
                 color='performance_level',
                 color_discrete_sequence=px.colors.qualitative.Set2,
                 title='Scatter Plot of Attendance vs Practice Score')

fig2.show()
```

```python
fig3 = px.box(df, x='performance_level', y='study_hours',
             title = 'Box Plot of Performance Level vs Study Hours')
fig3
```

```python
fig4 =px.histogram(df, x='practice_score', color='performance_level',
                   barmode='overlay',
             title = 'Histogram of Performance Level vs Practice Score')
fig4.show()
```

```python
corr=df.drop('student_id', axis = 'columns').corr(numeric_only= True)
```

```python
corr
```

```python
fig5 = px.imshow(corr,text_auto= True,
                 title='Correlation Heatmap')
fig5.show()
```
